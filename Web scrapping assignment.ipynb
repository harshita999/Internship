Write a python program to display IMDB’s 100 Top rated movies’ data (i.e. name, rating, year of release) and make data frame

```python
page1= requests.get('https://www.imdb.com/list/ls091520106/')
page1
```

```python
soup =BeautifulSoup(page1.text,'lxml')
```

```python
title= []   #empty for storing the list
for i in soup.find_all('h3',class_="lister-item-header"):
    title.append(i.text)
title
    
```

```python
rating=[]
for i in soup.find_all('div',class_='ipl-rating-star small'):
    rating.append(i.text)
rating
```

```python
release_years= []

for i in soup.find_all('span',class_="lister-item-year text-muted unbold"):
    release_years.append(i.text)
release_years
```

```python
#printing length
print(len(title),len(rating),len(release_years))
```

```python
#making dataframe
import pandas as pd
df= pd.DataFrame({'Titles':title, 'Ratings':rating,'Year of Release':release_years})
df
```
